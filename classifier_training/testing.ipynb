Testing to see if the correct (1,1) region is found.

```python
import matplotlib.pyplot as plt

# 1. load the img wanted
img_path = r"C:\Users\letit\Documents\GitHub\AutoTune-QD\classifier_training\images\005_alpha0.219_wobble37.3.png"
img = plt.imread(img_path)

# 2. these are the coords from my json
target_x = 301  # col
target_y = 197  # row

# 3. plotting 
fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(img, cmap="inferno") # default imshow puts 0,0 at top-left
ax.scatter(target_x, target_y, color="lime", s=100, marker="X", zorder=5)

ax.set_title(f"Target (1,1) at X={target_x}, Y={target_y}")
plt.show()
```

Extracting a patch from the target (1,1) region.

```python

# 1. calculate boundaries for a determined square
patch_size = 85 #45 was too small -> almost bblack
half = patch_size // 2

y1 = max(0, target_y - half)
y2 = min(img.shape[0], target_y + half + 1)
x1 = max(0, target_x - half)
x2 = min(img.shape[1], target_x + half + 1)

# 2. slice numpy array to extract the patch
patch = img[y1:y2, x1:x2]

# 3. plot  the zoomed in patch, with 4 rgba channels (a=alpha, the transparency) -> wont be used for svm tho
fig, ax = plt.subplots(figsize=(4, 4))
ax.imshow(patch, cmap="inferno")
ax.set_title(f"Extracted Patch\nShape: {patch.shape}")
ax.axis('off') 

plt.show()
```
